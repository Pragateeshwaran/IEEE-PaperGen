```python
from heapq import heappush, heappop

def heuristic(state, goal):
    return sum(abs(state.index(i) // 3 - goal.index(i) // 3) + abs(state.index(i) % 3 - goal.index(i) % 3) for i in range(1, 9) if i in state)

def get_neighbors(state):
    neighbors = []
    zero_index = state.index(0)
    row, col = zero_index // 3, zero_index % 3

    for dr, dc in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
        new_row, new_col = row + dr, col + dc
        if 0 <= new_row < 3 and 0 <= new_col < 3:
            new_state = list(state)
            new_state[zero_index], new_state[new_row * 3 + new_col] = new_state[new_row * 3 + new_col], 0
            neighbors.append(tuple(new_state))

    return neighbors

def a_star(initial, goal):
    heap = [(heuristic(initial, goal), initial, 0)]
    visited = set()

    while heap:
        _, state, cost = heappop(heap)
        if state == goal:
            return cost

        visited.add(state)

        for neighbor in get_neighbors(state):
            if neighbor not in visited:
                heappush(heap, (cost + 1 + heuristic(neighbor, goal), neighbor, cost + 1))

    return None

def main():
    initial =  (1, 3, 6, 5, 0, 2, 4, 7, 8)
    goal = (1, 2, 3, 4, 5, 6, 7, 8, 0)

    cost = a_star(initial, goal)
    if cost is not None:
        print(f"Solution found with cost: {cost}")
    else:
        print("No solution found.")

if __name__ == "__main__":
    main()
```

```python
def find_empty(board):
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                return i, j
    return None, None

def valid(board, num, pos):
    # Check row
    if num in board[pos[0]]:
        return False

    # Check column
    if num in [board[i][pos[1]] for i in range(9)]:
        return False

    # Check 3x3 box
    box_x = pos[1] // 3
    box_y = pos[0] // 3
    for i in range(box_y * 3, box_y * 3 + 3):
        for j in range(box_x * 3, box_x * 3 + 3):
            if board[i][j] == num:
                return False

    return True

def solve(board):
    row, col = find_empty(board)
    if row is None:
        return True  # No empty cells, the board is solved

    for num in range(1, 10):
        if valid(board, num, (row, col)):
            board[row][col] = num

            if solve(board):
                return True

            board[row][col] = 0

    return False

def print_board(board):
    for i in range(9):
        if i % 3 == 0 and i != 0:
            print("- - - - - - - - - - - -")

        for j in range(9):
            if j % 3 == 0 and j != 0:
                print(" | ", end="")

            if j == 8:
                print(board[i][j])
            else:
                print(str(board[i][j]) + " ", end="")

# Example usage
example_board = [
    [0, 0, 0, 2, 6, 0, 7, 0, 1],
    [6, 8, 0, 0, 7, 0, 0, 9, 0],
    [1, 9, 0, 0, 0, 4, 5, 0, 0],
    [8, 2, 0, 1, 0, 0, 0, 4, 0],
    [0, 0, 4, 6, 0, 2, 9, 0, 0],
    [0, 5, 0, 0, 0, 3, 0, 2, 8],
    [0, 0, 9, 3, 0, 0, 0, 7, 4],
    [0, 4, 0, 0, 5, 0, 0, 3, 6],
    [7, 0, 3, 0, 1, 8, 0, 0, 0]
]

print("Initial board:")
print_board(example_board)

if solve(example_board):
    print("\nSolved board:")
    print_board(example_board)
else:
    print("No solution exists for this Sudoku puzzle.")
```

```python
import heapq

class Graph:
    def __init__(self):
        self.edges = {}
        self.weights = {}
    
    def add_edge(self, from_node, to_node, weight):
        if from_node not in self.edges:
            self.edges[from_node] = []
        if to_node not in self.edges:
            self.edges[to_node] = []
        self.edges[from_node].append(to_node)
        self.edges[to_node].append(from_node)
        self.weights[(from_node, to_node)] = weight
        self.weights[(to_node, from_node)] = weight

def ucs(graph, start, goal):
    # Priority queue to store (cost, node, path)
    queue = [(0, start, [])]
    # Set to store visited nodes
    visited = set()
    
    while queue:
        (cost, node, path) = heapq.heappop(queue)
        
        if node not in visited:
            visited.add(node)
            path = path + [node]
            
            if node == goal:
                return (cost, path)
            
            for neighbor in graph.edges[node]:
                if neighbor not in visited:
                    total_cost = cost + graph.weights[(node, neighbor)]
                    heapq.heappush(queue, (total_cost, neighbor, path))
    
    return (float('inf'), [])  # If goal not reachable

# Example usage
graph = Graph()
graph.add_edge('A', 'B', 1)
graph.add_edge('A', 'C', 4)
graph.add_edge('B', 'C', 2)
graph.add_edge('B', 'D', 5)
graph.add_edge('C', 'D', 1)

start = 'A'
goal = 'D'
cost, path = ucs(graph, start, goal)
print(f"Cost: {cost}, Path: {path}")
```

```python
from collections import deque

class Graph:
    def __init__(self):
        self.edges = {}
    
    def add_edge(self, from_node, to_node):
        if from_node not in self.edges:
            self.edges[from_node] = []
        if to_node not in self.edges:
            self.edges[to_node] = []
        self.edges[from_node].append(to_node)
        self.edges[to_node].append(from_node)

def bfs_step(queue, visited_from_this_side, visited_from_other_side, parents):
    current_node, path = queue.popleft()

    for neighbor in graph.edges[current_node]:
        if neighbor in visited_from_other_side:
            return path + [neighbor]
        
        if neighbor not in visited_from_this_side:
            visited_from_this_side.add(neighbor)
            parents[neighbor] = current_node
            queue.append((neighbor, path + [neighbor]))
    
    return None

def construct_path(parents_from_start, parents_from_goal, meeting_node):
    path_from_start = []
    node = meeting_node
    while node is not None:
        path_from_start.append(node)
        node = parents_from_start.get(node, None)
    path_from_start.reverse()

    path_from_goal = []
    node = meeting_node
    while node is not None:
        path_from_goal.append(node)
        node = parents_from_goal.get(node, None)

    return path_from_start + path_from_goal[1:]

def bidirectional_search(graph, start, goal):
    if start == goal:
        return [start]

    queue_start = deque([(start, [start])])
    queue_goal = deque([(goal, [goal])])

    visited_start = set([start])
    visited_goal = set([goal])

    parents_from_start = {start: None}
    parents_from_goal = {goal: None}

    while queue_start and queue_goal:
        path = bfs_step(queue_start, visited_start, visited_goal, parents_from_start)
        if path:
            meeting_node = path[-1]
            return construct_path(parents_from_start, parents_from_goal, meeting_node)
        
        path = bfs_step(queue_goal, visited_goal, visited_start, parents_from_goal)
        if path:
            meeting_node = path[-1]
            return construct_path(parents_from_start, parents_from_goal, meeting_node)

    return None  # If no path is found

# Example usage
graph = Graph()
graph.add_edge('A', 'B')
graph.add_edge('A', 'C')
graph.add_edge('B', 'C')
graph.add_edge('B', 'D')
graph.add_edge('C', 'D')
graph.add_edge('C', 'E')
graph.add_edge('D', 'E')

start = 'A'
goal = 'E'
path = bidirectional_search(graph, start, goal)
print(f"Path: {path}")
```

```python
from collections import deque

class State:
    def __init__(self, a, b):
        self.a = a
        self.b = b

def water_jug_bfs(a, b, x):
    start_state = State(0, 0)
    visited = set()
    queue = deque([(start_state, [])])

    while queue:
        current_state, path = queue.popleft()

        if current_state.a == x or current_state.b == x:
            return path

        visited.add(current_state)

        actions = [
            (a, current_state.b, 'Fill A'),
            (current_state.a, b, 'Fill B'),
            (0, current_state.b, 'Empty A'),
            (current_state.a, 0, 'Empty B'),
            (max(0, current_state.a - (b - current_state.b)), min(b, current_state.b + current_state.a), 'Pour A to B'),
            (min(a, current_state.a + current_state.b), max(0, current_state.b - (a - current_state.a)), 'Pour B to A')
        ]

        for action in actions:
            new_state = State(action[0], action[1])
            if new_state not in visited:
                queue.append((new_state, path + [action[2]]))

    return None

# Example usage
a_capacity = 4  # Capacity of jug A
b_capacity = 3  # Capacity of jug B
target = 2      # Target amount of water

path = water_jug_bfs(a_capacity, b_capacity, target)
if path:
    print("Steps to get {} liters of water into the A liter jug:".format(target))
    for step in path:
        print(step)
else:
    print("Cannot achieve the target amount of water.")
```
